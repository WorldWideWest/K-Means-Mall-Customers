```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
data = pd.read_csv('source\datasets_177976_401124_Mall_Customers.csv')
data.head()
```

```python
# maping genders 1-Female, 0-Male
data['Genre'] = data['Genre'].map({'Female': 1, 'Male': 0}) 
data.head()
```

```python
dataF = data[data['Genre'] == 1].count()
dataM = data[data['Genre'] == 0].count()
genderData = pd.DataFrame({'Gender': ['Female', 'Male'],
                           'count': [dataF[0], dataM[0]]})

plt.bar(x = genderData['Gender'],
        height = genderData['count'],
        color = ['#D74E09', '#6E0E0A'],
        width = 0.4,
        align = 'center')

plt.title('Number of participants in the survey, devided into Genders',
          fontsize = 14,
          fontweight = 'bold')
plt.xlabel('Gender',
          fontsize = 12,
          fontweight = 'bold')
plt.ylabel('Number of participants',
          fontsize = 12,
          fontweight = 'bold')
plt.savefig('visualizations/GENDER.png', dpi = 300)
plt.show()
```

```python
plt.hist(x = data['Age'],
         bins = 7,
         histtype = 'bar',
         color = '#124E78')

plt.title('Age distribution in the survey',
          fontsize = 14,
          fontweight = 'bold')
plt.xlabel('Years',
           fontsize = 12,
           fontweight = 'bold')
plt.ylabel('Number of participants',
           fontsize = 12,
           fontweight = 'bold')
plt.savefig('visualizations/ADS.png', dpi = 300)
plt.show()
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(1,3, figsize = (18,5))

ax1.scatter(data['Annual Income (k$)'], data['Spending Score (1-100)'],                               alpha = 0.5,
            color = '#FF3864')
ax1.set_title('Annual Income compared to Spending Score',
              fontsize = 12,
              fontweight = 'bold')
ax1.set_xlabel('Annual Income',
               fontsize = 8,
               fontweight = 'bold')
ax1.set_ylabel('Spending Score',
               fontsize = 8,
               fontweight = 'bold')

ax2.scatter(data['Age'], data['Annual Income (k$)'],
            alpha = 0.5,
            color = '#261447')
ax2.set_title('Annual Income compared to Age',
              fontsize = 12,
              fontweight = 'bold')
ax2.set_xlabel('Age',
               fontsize = 8,
               fontweight = 'bold')
ax2.set_ylabel('Annual Income',
               fontsize = 8,
               fontweight = 'bold')

ax3.scatter( data['Age'], data['Spending Score (1-100)'],
            alpha = 0.5,
            color = '#F39C6B')
ax3.set_title('Spending Score compared to Age',
              fontsize = 12,
              fontweight = 'bold')
ax3.set_xlabel('Age',
               fontsize = 8,
               fontweight = 'bold')
ax3.set_ylabel('Spending Score',
               fontsize = 8,
               fontweight = 'bold')
plt.savefig('visualizations/AISAIASSA.png', dpi = 300)
plt.show()
```
